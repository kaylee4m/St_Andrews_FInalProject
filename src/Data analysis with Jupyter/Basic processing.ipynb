```python
import json
import xlwt
import pandas as pd
from pandas.io.json import json_normalize
import os
```

```python
f = open('perdata.json', encoding='utf-8')
info = json.load(f)
 
wb = xlwt.Workbook()
sh = wb.add_sheet('Sheet 3')
```

```python
# Initialization data
attributes = []
 
for x in range(len(info["items"])):
    for attr in info["items"][x].keys():
        if attr not in attributes:
            print(attr)
            attributes.append(attr)
            sh.write(0, attributes.index(attr), attr)
        sh.write(x+1, attributes.index(attr), str(info["items"][x][attr]))

##wb.save('XXXXXXX.xls')
```
